### In dieser Aufgabe erzeugen wir ein Sinussignal und ein Cosinussignal mit je 200 Samples über einen Zeitraum von 1 s und plotten beide gemeinsam.

Wir verwenden ein **numpy-Array** zum Abspeichern der Zeitmarken *t* und der Funktionswerte. numpy-Arrays werden mit *"import numpy as np"* importiert.

Zum Plotten verwenden wir die mittels **pip** installierte Bibliothek **matplotlib**, aus der wir das Modul *pyplot* als *plt* importieren. Mit **plt.style.use()** kann man verschiedene Plot-Styles verwenden, siehe [Matplotlib Style Sheets](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

Für den **Betrag** komplexer Zahlen (z. B. des komplexen Signals $z$) wird **np.abs()** verwendet: `np.abs(z)` liefert den Betrag (Magnitude) für reelle und komplexe Arrays elementweise.


```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')

f = 5.0
t = np.linspace(0, 1, 200, endpoint=False)
y_sin = np.sin(2 * np.pi * f * t)
y_cos = np.cos(2 * np.pi * f * t)
```

```python
plt.plot(t, y_cos, label='cos')
plt.plot(t, y_sin, label='sin')
plt.xlabel('t / s')
plt.legend()
plt.show()
```

### Komplexes Signal: Cosinus und Sinus als Komponenten

Cosinus (Realteil) und Sinus (Imaginärteil) lassen sich zu einem **komplexen Signal** zusammenfassen:

$$z(t) = \cos(2\pi f t) + j\,\sin(2\pi f t) = e^{j2\pi f t}$$

oder:

$$z(t) = \cos(\omega t) + j\,\sin(\omega t) = e^{j\omega t}$$

Dabei ist $j$ die imaginäre Einheit. Der **Realteil** von $z$ ist der Cosinus, der **Imaginärteil** der Sinus. In der komplexen Ebene beschreibt $z(t)$ einen Kreis mit Radius 1.

```python
# Komplexes Signal: z = cos + j*sin
z = y_cos + 1j * y_sin
# Absolutbetrag (Betrag der komplexen Zahl, hier für alle Zeitpunkte)
magnitude = np.abs(z)
print("Absolutbetrag (erste 5 Werte):", magnitude[:5])
print("Erwartung (konstant 1): alle Werte ≈ 1.0")
plt.figure()
plt.plot(t, magnitude)
plt.xlabel('t / s')
plt.ylabel('|z|')
plt.title('Absolutbetrag des komplexen Signals z = cos + j·sin')
plt.show()
```

## Aufgabe 1

**Ersetze** die `np.sin()`-Funktion durch eine `np.cos()`-Funktion mit entsprechender **Phasenverschiebung**, sodass daraus wieder eine Sinusfunktion entsteht.

*Hinweis:* Es gilt $\sin(x) = \cos(x - \pi/2)$. In NumPy: `np.pi` für $\pi$.

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')

f = 5.0
t = np.linspace(0, 1, 200, endpoint=False)

# ergänze die Berechnung für y_sin und y_cos
y_sin = np.sin(2 * np.pi * f * t)
y_cos = np.cos(2 * np.pi * f * t - np.pi / 2)
```

```python
plt.plot(t, y_cos, label='cos')
plt.plot(t, y_sin, label='sin')
plt.xlabel('t / s')
plt.legend()
plt.show()
```
